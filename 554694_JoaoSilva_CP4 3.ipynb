## **Moda, média, mediana, variância e desvio padrão:**

```python
# Selecionar apenas as colunas numéricas
numeric_columns = df_titanic.select_dtypes(include=['number'])

# Concatenar todas as colunas numéricas em uma única série
all_data_combined = pd.concat([numeric_columns[col] for col in numeric_columns.columns])

# Calcular a moda
moda_combined = all_data_combined.mode().iloc[0]

# Calcular a média
media_combined = all_data_combined.mean()

# Calcular a mediana
mediana_combined = all_data_combined.median()

# Calcular a variância
variancia_combined = all_data_combined.var()

# Calcular o desvio padrão
desvio_padrao_combined = all_data_combined.std()

# Exibir os resultados
estatisticas_combined = {
    'Moda': moda_combined,
    'Média': media_combined,
    'Mediana': mediana_combined,
    'Variância': variancia_combined,
    'Desvio Padrão': desvio_padrao_combined
}

for key, value in estatisticas_combined.items():
    print(f'{key}: {value:.2f}')
```

## **Qual é a moda, a média, a mediana, a variância e o desvio padrão das colunas Fare e Age?**

```python
columns_of_interest = ['Fare', 'Age']

for column in columns_of_interest:
    print(f"\nAnálise para a coluna {column}:")
    print(f"Moda: {df_titanic[column].mode()[0]:.2f}")
    print(f"Média: {df_titanic[column].mean():.2f}")
    print(f"Mediana: {df_titanic[column].median():.2f}")
    print(f"Variância: {df_titanic[column].var():.2f}")
    print(f"Desvio Padrão: {df_titanic[column].std():.2f}")
```

# **Gráficos:**

### **Demonstre com um gráfico a quantidade de sobreviventes e não sobreviventes de acordo com classe de embarque.**

```python
# Carregar o dataset
df_titanic = pd.read_csv("/content/drive/MyDrive/Colab/Titanic-Dataset.csv")

# Gráfico de sobreviventes e não sobreviventes por classe
plt.figure(figsize=(10, 6))
sns.countplot(data=df_titanic, x='Pclass', hue='Survived')
plt.title('Quantidade de Sobreviventes e Não Sobreviventes de Acordo com a Classe de Embarque')
plt.xlabel('Classe de Embarque')
plt.ylabel('Quantidade')
plt.legend(title='Sobrevivente')
plt.show()
```

**Minha interpretação:** Como mostra o gráfico, a classe de embarque desempenhou um papel significativo na sobrevivência dos passageiros do Titanic. Os passageiros das classes superiores (primeira e segunda classe) tiveram maiores chances de sobreviver, enquanto os passageiros da terceira classe tiveram as piores chances. Isso mostra as desigualdades sociais daquela época, pois os mais ricos tinham mais recursos para emergências.

## **Demonstre com um gráfico a frequência de Age.**
